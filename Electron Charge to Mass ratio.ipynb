## Background

Electron charge to mass ratio experiment was done by Schuster in 1890. By bending a beam of electrons with a magnetic field and measuring the radius of the electron's path and the strength of the field, he was able to determine the charge to mass ratio of a negative particle.

## Theory

The charge to mass ratio of an electron is a constant. By putting a beam of electrons that have been accelerated by an electric field in a uniform magnetic field created by Helmholtz coils, the beam will bend along a circular path. Measuring the radius of this path will allow us to determine the electron charge to mass ratio with the equation.

$$\frac{e}{m}=\frac{2V}{B^2R^2}$$

Where *V* is the potential difference of the electric field used to accelerate the electrons, *B* is the strength of the magnetic field created by the Helmholtz coils, and *R* is the radius of the electron beam's circular path. 

## Procedure

In our experiment we will be using the CENCO e/m apparatus to determine the radius. The apparatus consists of a beam of electrons created by accelerating electrons from a cathode through an electric field beneath an etched disk. This beam is curved by a magnetic field created by Helmholz coils so that it goes through a hole in the center of the disk, at which point the electric field ends, and hits one of the etched rings in the disk. Using the cathode as the origin, the disk is at a known height *d*=0.254 cm and the etched rings are at known distances *X* from the center of the disk. By varying the current going through the Helmholz coils that are creating the magnetic field, we can make the beam hit the etched rings and measure the corresponding values for *R* with the equation

$$R=\frac{X}{2}+\frac{d^2}{3X}$$



## Data Analysis

The values for *R*,*B*, and *V* can be plugged into the equation for the charge to mass ratio directly, or a linear least squares regression fit can be done on the data. By plotting *I* vs. 1/*R*, where *I* is the current through the Helmholtz coils, we can obtain a slope that will let us solve for the charge to mass ratio.
